```sql
-- Create external table with partitioning by year extracted from shipdate
CREATE EXTERNAL TABLE IF NOT EXISTS hive_metastore.default.lineitem_partitioned_hive (
  l_orderkey BIGINT,
  l_partkey BIGINT,
  l_suppkey BIGINT,
  l_linenumber INT,
  l_quantity DECIMAL(15,2),
  l_extendedprice DECIMAL(15,2),
  l_discount DECIMAL(15,2),
  l_tax DECIMAL(15,2),
  l_returnflag STRING,
  l_linestatus STRING,
  l_shipdate DATE,
  l_commitdate DATE,
  l_receiptdate DATE,
  l_shipmode STRING,
  l_comment STRING
)
USING DELTA
PARTITIONED BY (year INT)
LOCATION 's3://demonhid15-rootbucket-sm/external_source_ext/lineitem_partitioned_hive';
```
